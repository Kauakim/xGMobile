 ### 3. Regressão Polinomial – Exercícios

**Exercício 5:**
Considere o seguinte modelo gerador de dados:

$$
y = 2x^2 - x + 1 + w,
$$

onde:

- $x$ é um vetor coluna com $N = 1000$ amostras geradas de forma independente a partir de uma distribuição uniforme no intervalo $[-1, 1]$;
- $w$ é um vetor coluna com $N$ amostras extraídas de uma distribuição normal Gaussiana com média zero e desvio padrão $\sigma = 0.5$.

O objetivo é analisar o comportamento da regressão polinomial ao tentar aproximar um modelo não linear verdadeiro utilizando diferentes graus de complexidade.
Treine modelos de regressão polinomial considerando as seguintes hipóteses:

- Grau 1 (modelo linear):

$$
y \approx \beta_0 + \beta_1 x
$$

- Grau 2 (modelo correto em relação ao processo gerador):

$$
y \approx \beta_0 + \beta_1 x + \beta_2 x^2
$$

- Grau 3 (modelo superparametrizado):

$$
y \approx \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3
$$

**Objetivo:**

Comparar o desempenho dos três modelos em termos de:

- capacidade de aproximação da função verdadeira;
- plotar a função original, os dados ruidosos e a reta estimada.
- impacto da complexidade do modelo (underfitting vs. overfitting);
- influência do termo de ruído $w$ na estimativa dos coeficientes.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
# Exercício 1

# 1. Gerando os dados
N = 1000                                  # Número de amostras
x = np.linspace(-1, 1, N).reshape(N, 1)   # Entrada (Nx1)
w = np.sqrt(0.1) * np.random.randn(N, 1)  # Ruído gaussiano
y = 3 * x + 2                             # Função original
y_noisy = y + w                           # Função ruidosa

# 2. Criando o modelo de Regressão Linear
model = LinearRegression()

# 3. Treinando o modelo com os dados ruidosos
model.fit(x, y_noisy)

# 4. Obtendo parâmetros estimados
a_hat = model.coef_[0][0]       # Inclinação
b_hat = model.intercept_[0]     # Intercepto

print(f"Coeficiente estimado (â): {a_hat:.3f}")
print(f"Intercepto estimado (b̂): {b_hat:.3f}")

# 5. Predição nos mesmos pontos
y_pred = model.predict(x)

# 6. Visualização do resultado
plt.figure(figsize=(8,5))
plt.scatter(x, y_noisy, s=8, label="Amostras com ruído")
plt.plot(x, y, 'k--', label="Função original (y = -1 + 2x)")
plt.plot(x, y_pred, 'r', linewidth=2, label="Reta ajustada pelo modelo")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão Linear Simples — Ajuste da Reta")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercício 2

# 1. Gerando os dados
N = 100                                   # Número de amostras
x = np.linspace(0, 10, N).reshape(N, 1)   # Entrada (Nx1)
w = 2000 * np.random.randn(N, 1)          # Ruído gaussiano
y = 50000 -3000 * x                   # Função original
y_noisy = y + w                           # Função ruidosa

# 2. Criando o modelo de Regressão Linear
model = LinearRegression()

# 3. Treinando o modelo com os dados ruidosos
model.fit(x, y_noisy)

# 4. Obtendo parâmetros estimados
beta1_hat = model.coef_[0][0]   # Inclinação
beta0_hat = model.intercept_[0] # Intercepto

print(f"Coeficiente estimado (β̂1): {beta1_hat:.3f}")
print(f"Intercepto estimado (β̂0): {beta0_hat:.3f}")

# 5. Predição nos mesmos pontos
y_pred = model.predict(x)

# 6. Visualização do resultado
plt.figure(figsize=(8,5))
plt.scatter(x, y_noisy, s=8, label="Amostras com ruído")
plt.plot(x, y, 'k--', label="Função original (y = 50000 -3000x)")
plt.plot(x, y_pred, 'r', linewidth=2, label="Reta ajustada pelo modelo")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão Linear Simples — Ajuste da Reta")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercício 3

# 1. Gerando os dados
N = 1000

# Duas variáveis de entrada
x1 = np.random.uniform(-1, 1, N) # (Nx1)
x2 = np.random.uniform(-1, 1, N)

# Junta em uma matriz (N x 2)
X = np.column_stack((x1, x2))

# Função original
y = 5 + 3*x1 - 2*x2

# Ruído
w = 0.5 * np.random.randn(N)
y_noisy = y + w

# 2. Modelo
model = LinearRegression()

# 3. Treinamento
model.fit(X, y_noisy)

# 4. Parâmetros estimados
b_hat = model.intercept_
a_hat = model.coef_

print("Intercepto estimado (b̂):", b_hat)
print("Coeficientes estimados (â):", a_hat)

# 5. Predição
y_pred = model.predict(X)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y_noisy, alpha=0.3, label="Dados reais (ruído)")

# Criando um grid para o plano
x1_grid, x2_grid = np.meshgrid(
    np.linspace(min(x1), max(x1), 40),
    np.linspace(min(x2), max(x2), 40))

# Plano previsto
y_grid = (b_hat
    + a_hat[0] * x1_grid
    + a_hat[1] * x2_grid)

# Superfície do plano
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5, color='red')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.set_title("Regressão Linear Múltipla")
plt.legend()
plt.show()
```

```python
# Exercício 4

# 1. Gerando os dados
N = 1000

# Duas variáveis de entrada
x1 = np.random.uniform(0, 10, N) # (Nx1)
x2 = np.random.uniform(0, 10000, N)

# Junta em uma matriz (N x 2)
X = np.column_stack((x1, x2))

# Normaliza a matriz
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Extrai os eixos x1 e x2 já normalizados
x1_normalized = X_normalized[:, 0]
x2_normalized = X_normalized[:, 1]

# Função original
y = 4*x1 + 0.002*x2

# Ruído
w = 5 * np.random.randn(N)
y_noisy = y + w

# 2. Modelo
model1 = LinearRegression()
model2 = LinearRegression()

# 3. Treinamento
model1.fit(X, y_noisy)
model2.fit(X_normalized, y_noisy)

# 4. Parâmetros estimados
b_hat = model1.intercept_
a_hat = model1.coef_
b_hat_normalized = model2.intercept_
a_hat_normalized = model2.coef_

score1 = model1.score(X, y_noisy)
score2 = model2.score(X_normalized, y_noisy)

print("--- SEM NORMALIZAÇÃO ---")
print("Intercepto (b̂):", b_hat)
print("Coeficientes (â):", a_hat)
print("Score (R²):", score1, "\n")

print("--- COM NORMALIZAÇÃO ---")
print("Intercepto (b̂):", b_hat_normalized)
print("Coeficientes (â):", a_hat_normalized)
print("Score (R²):", score2)

print("PQ A NORMALIZACAO FOI MELHOR")

# 5. Predição
# Grid sem normalização
x1_grid, x2_grid = np.meshgrid(
    np.linspace(min(x1), max(x1), 40),
    np.linspace(min(x2), max(x2), 40))

pontos_grid1 = np.column_stack((x1_grid.ravel(), x2_grid.ravel()))
y_grid1 = model1.predict(pontos_grid1).reshape(x1_grid.shape)

# Grid com normalização
x1_norm_grid, x2_norm_grid = np.meshgrid(
    np.linspace(min(x1_normalized), max(x1_normalized), 40),
    np.linspace(min(x2_normalized), max(x2_normalized), 40))

pontos_grid2 = np.column_stack((x1_norm_grid.ravel(), x2_norm_grid.ravel()))
y_grid2 = model2.predict(pontos_grid2).reshape(x1_norm_grid.shape)

# Gráfico 1 (Sem normalização)
fig1 = plt.figure(figsize=(10, 7))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.scatter(x1, x2, y_noisy, alpha=0.3, label="Dados reais (ruído)")
ax1.plot_surface(x1_grid, x2_grid, y_grid1, alpha=0.5, color='red')
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_zlabel("y")
ax1.set_title(f"Regressão Sem Normalização\nScore (R²): {score1:.4f}")
ax1.legend()
plt.show()

# Gráfico 2 (Com normalização)
fig2 = plt.figure(figsize=(10, 7))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(x1_normalized, x2_normalized, y_noisy, alpha=0.3, label="Dados normalizados")
ax2.plot_surface(x1_norm_grid, x2_norm_grid, y_grid2, alpha=0.5, color='blue')
ax2.set_xlabel("x1 (Normalizado)")
ax2.set_ylabel("x2 (Normalizado)")
ax2.set_zlabel("y")
ax2.set_title(f"Regressão Com Normalização\nScore (R²): {score2:.4f}")
ax2.legend()
plt.show()
```

```python
# Exercício 5

# 1. Dados artificiais
N = 1000
x = np.linspace(-1, 1, N).reshape(-1, 1)

# Ruído
w = 0.5 * np.random.randn(N, 1)

# Função original
y = 2*(x**2) - x + 1

# Dados com ruído
y_noisy = y + w

# 2. Transformação polinomial
poly2 = PolynomialFeatures(degree=2, include_bias=True)
poly3 = PolynomialFeatures(degree=3, include_bias=True)
X_poly2 = poly2.fit_transform(x)   # vira [1, x, x²]
X_poly3 = poly3.fit_transform(x)   # vira [1, x, x², x³]

# 3. Treinamento
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
model1.fit(x, y_noisy)
model2.fit(X_poly2, y_noisy)
model3.fit(X_poly3, y_noisy)

# 4. Parâmetros estimados
b_hat1 = model1.intercept_
a_hat1 = model1.coef_
b_hat2 = model2.intercept_
a_hat2 = model2.coef_
b_hat3 = model3.intercept_
a_hat3 = model3.coef_

score1 = model1.score(x, y_noisy)
score2 = model2.score(X_poly2, y_noisy)
score3 = model3.score(X_poly3, y_noisy)

print("--- Grau 1 ---")
print("Intercepto (b̂):", b_hat1)
print("Coeficientes (â):", a_hat1)
print("Score (R²):", score1, "\n")

print("--- Grau 2 ---")
print("Intercepto (b̂):", b_hat2)
print("Coeficientes (â):", a_hat2)
print("Score (R²):", score2)

print("--- Grau 2 ---")
print("Intercepto (b̂):", b_hat3)
print("Coeficientes (â):", a_hat3)
print("Score (R²):", score3)

# Análise dos resultados
print("Como visível pelos valores obtidos dos coeficientes e dos scores dos modelos, é possivel perceber os efeitos de underfitting, com o modelo de grau 1, e overfitting, com o modelo de grau 3, já que o problema exige um modelo de grau 2.")
print("No modelo de grau 1, a predição gerada é simples demais, enquanto no modelo de grau 3 é complexa demais")

# 5. Predição
y_pred1 = model1.predict(x)
y_pred2 = model2.predict(X_poly2)
y_pred3 = model3.predict(X_poly3)

# 6. Plot
plt.figure(figsize=(10,6))
plt.scatter(x, y_noisy, alpha=0.15, label="Dados ruidosos")
plt.plot(x, y, label="Função original", linewidth=2)
plt.plot(x, y_pred1, label="Regressão Polinomial (grau 1 - Underfitting)", linewidth=3)
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(color='black')
plt.axhline(color='black', linestyle="--")
plt.title("Regressão Polinomial (grau 1 - Underfitting)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10,6))
plt.scatter(x, y_noisy, alpha=0.15, label="Dados ruidosos")
plt.plot(x, y, label="Função original", linewidth=2)
plt.plot(x, y_pred2, label="Regressão polinomial (grau 2 - Ideal)", linewidth=3)
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(color='black')
plt.axhline(color='black', linestyle="--")
plt.title("Regressão polinomial (grau 2 - Ideal)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10,6))
plt.scatter(x, y_noisy, alpha=0.15, label="Dados ruidosos")
plt.plot(x, y, label="Função original", linewidth=2)
plt.plot(x, y_pred3, label="Regressão polinomial (grau 3 - Overfitting)", linewidth=3)
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(color='black')
plt.axhline(color='black', linestyle="--")
plt.title("Regressão polinomial (grau 3 - Overfitting)")
plt.legend()
plt.grid(True)
plt.show()
```
